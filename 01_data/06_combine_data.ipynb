## Split datasets to train, validation and test sets

```python
def generate_split_qa_pairs(data):
    '''
    Purpose: to put all data into proper question-answer pair JSON format
    @params data: loaded json data with augmented (not in QA pair json format) data from json file (list of dicts)
    returns: 3 lists of json-formatted question-answer pairs into a train, validation and test set (list of dicts)
    '''
    train_set = []
    val_set = []
    test_set = []
    for dat in data:
        
        # put original & augmented questions into a list
        questions = [dat['original_question']]  # total of 1 question; this is the original question
        questions += dat['augment_questions']   # total of 7 variations of the original question
        
        # randomly sample questions to split for each data point/question
        train = random.sample(questions, 6)                               # sample 6 variations of the question for train set
        val = random.sample([x for x in questions if x not in train], 1)  # sample 1 variation of the question (of remaining variaons) for validation set
        test = [x for x in questions if x not in train and x not in val]  # use remaining 1 variation of question for test set
        
        # pair all questions with the answer
        ans = dat['new_answer']
        
        train_qa = [{'question':q, 'answer':ans} for q in train]
        val_qa = [{'question':q, 'answer':ans} for q in val]
        test_qa = [{'question':q, 'answer':ans} for q in test]
        
        train_set += train_qa
        val_set += val_qa
        test_set += test_qa
    
    return train_set, val_set, test_set
```

```python
def save_data_to_file(data, dir_path, savefilename):
    '''
    Purpose: to save data to a JSON file
    @params data: the new data to save (list of dicts)
    @params dir_path: path to directory to save data file in (str)
    @params savefilename: the name of file to save data in (str)
    returns: success message (str)
    '''
    if os.path.isfile(f'{dir_path}/{savefilename}') == False:  # if file to save new data doesn't exist already
        with open(f'{dir_path}/{savefilename}', 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        f.close()

    else:  # if file to save new data does exist already
        # read/load existing data
        f = open(f'{dir_path}/{savefilename}')
        existing_data = json.load(f)
        f.close()

        # add new data
        existing_data += data

        # save data to file
        with open(f'{dir_path}/{savefilename}', 'w', encoding='utf-8') as f:
            json.dump(existing_data, f, ensure_ascii=False, indent=4)
        f.close()
        
    return f'[SUCCESS] All data is saved in file in: {dir_path}/{savefilename}'
```

```python
# input path of directory of where 'all_data.json' file is located -- this directory will store new data and error file as well
alldata_dir = '/Users/dahliama/Desktop/data298/github/projectmarley/data/data-preprocess/03_data_augments'

dahlia_file = 'dahlia_augment_data.json'
deepak_file = 'deepak_augment_data.json'
gabriel_file = 'gabriel_augment_data.json'
sungjun_file = 'sungjun_augment_data.json'
juyeon_file = 'juyeon_augment_data.json'

train_savefilename = 'train_augment_data.json'
val_savefilename = 'val_augment_data.json'
test_savefilename = 'test_augment_data.json'
```

```python
# create list of all files to load and split
load_files = [dahlia_file
              ,deepak_file
              ,gabriel_file
              ,sungjun_file
              ,juyeon_file]

for file in load_files:
    
    # load data subset file
    f = open(f'{alldata_dir}/{file}')
    data = json.load(f)
    f.close()
    
    print(f"{file} has {len(data)} data points.")
    
    # generate question-answer pairs and split data
    train_data, val_data, test_data = generate_split_qa_pairs(data)
    
    # save train dataset to file
    return_msg = save_data_to_file(train_data, alldata_dir, train_savefilename)
    return_msg = save_data_to_file(val_data, alldata_dir, val_savefilename)
    return_msg = save_data_to_file(test_data, alldata_dir, test_savefilename)
    
    # check total data count for each file
    total_data_len = len(train_data) + len(val_data) + len(test_data)
    print(f"{file} has a total of {total_data_len} data points after augmentation.\n")
```

```python
# shuffle all split datasets
load_files = [train_savefilename
              ,val_savefilename
              ,test_savefilename]

for file in load_files:
    
    # load data subset file
    f = open(f'{alldata_dir}/{file}')
    data = json.load(f)
    f.close()
    
    # shuffle data
    random.shuffle(data)
    
    # save shuffled data
    with open(f'{alldata_dir}/{file}', 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
    f.close()
```

```python
# input path of directory of where 'all_data.json' file is located -- this directory will store new data and error file as well
alldata_dir = '/Users/dahliama/Desktop/data298/github/projectmarley/data/data-preprocess/03_data_augments'

dahlia_file = 'dahlia_augment_data.json'
deepak_file = 'deepak_augment_data.json'
gabriel_file = 'gabriel_augment_data.json'
sungjun_file = 'sungjun_augment_data.json'
juyeon_file = 'juyeon_augment_data.json'

savefilename = 'all_augment_data.json'
```

```python
def generate_qa_pairs(data):
    '''
    Purpose: to put augmented data into proper question-answer pair JSON format
    @params data: loaded json data with augmented (not in QA pair json format) data from json file (list of dicts)
    returns: a list of json-formatted question-answer pairs (list of dicts)
    '''
    qa_pairs = []
    for dat in data:
        qa = {}
        qa['question'] = dat['original_question']
        qa['answer'] = dat['new_answer']
        qa_pairs.append(qa)

        for aug_q in dat['augment_questions']:
            qa = {}
            qa['question'] = aug_q
            qa['answer'] = dat['new_answer']
            qa_pairs.append(qa)
    
    return qa_pairs
```

```python
# load and combine data
load_files = [dahlia_file
              ,deepak_file
              ,gabriel_file
              ,sungjun_file
              ,juyeon_file]

all_data = []
for file in load_files:
    
    # load file
    f = open(f'{alldata_dir}/{file}')
    data = json.load(f)
    f.close()
    
    print(f"{file} has {len(data)} data points.")
    
    # generate question-answer pairs for each file
    qa_pairs = generate_qa_pairs(data)
    all_data += qa_pairs
    
    print(f"{file} has a total of {len(qa_pairs)} data points after augmentation.\n")
    
print(f"\n[COMPLETE] There is a total of {len(all_data)} data points.")
```

```python
# export and save combined data
with open(f'{alldata_dir}/{savefilename}', 'w', encoding='utf-8') as f:
    json.dump(all_data, f, ensure_ascii=False, indent=4)
f.close()
```
