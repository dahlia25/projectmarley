# Creating Question-Answer Pairs JSON data from raw text bodies with OpenAI GPT3.5-Turbo
This notebook contains testing and automated code to generate question-answer (QA) pairs JSON data from raw text with OpenAI's GPT3.5-turbo model.

## Define functions

```python
def get_response(api_key, prompt, max_tokens=None, temperature=0.7, model="gpt-3.5-turbo"):
    '''
    Purpose: to input raw text to OpenAI API GPT model to generate QA pairs in JSON format
    @params api_key: input API key from OpenAI (str)
    @params prompt: input prompt of what we want GPT model to do (str)
    @params max_tokens: input max tokens for model to generate completion; default set to None, since automatically calculated in function
    @params temperature: input randomness/creativity of model; default is 0.7 (which is also the default used by OpenAI)
    @params model: input OpenAI model to use; default is gpt-3.5-turbo
    returns: a response object from the API
    '''
    if max_tokens == None:
        max_tokens = round(4096 - (len(prompt)/4))  # calculate max_tokens for completion; divide prompt by 4 since 1 token ~= 4 chars
    else:  # override if max_tokens parameter is specified
        max_tokens = max_tokens
    
    client = openai.OpenAI(
        api_key=api_key
    )
    
    messages = [{"role": "user", "content": prompt}]
    
    response = client.chat.completions.create(
        messages=messages,
        model=model,
        max_tokens=max_tokens,
        temperature=temperature,
        stop=None # provide a list of strings to stop generation at certain points
    )

    return response
```

**Run code**

```python
folders = sorted([f for f in os.listdir(rawdata_dir) if '.' not in f])

counter = 0         # set counter to track number of API requests made
# break_ind = False   # an indicator to indicate if need to break out of loop due to hitting API rate limit
for folder in folders:
    print(f'----- Processing files in "{folder}" data folder -----')
    
    files = sorted([f for f in os.listdir(f'{rawdata_dir}/{folder}') if '.txt' in f])
    print(f'\t.. total # of files = {len(files)}')
    
    break_ind = False   # an indicator to indicate if need to break out of folder loop due to an error
    
    for i in tqdm(range(len(files))):
        try:
            if i % 10 == 0:
                time.sleep(2) # have code execution delay by 2 seconds for every 10 API requests made
            else:
                # # break out of loop if hit the API request rate limit
                # if counter == 200:  # 200 = API requests limit per day
                #     print(f'Hit API Rate Limit warning: made {counter} requests')
                #     break_ind == True
                #     break
                
                # load text file
                file = files[i]
                file_obj = open(f'{rawdata_dir}/{folder}/{file}', 'r')
                text = file_obj.read()
                file_obj.close()

                # generate question-answer pairs using OpenAI API
                prompt_prefix = 'Create as many question and answer pairs in JSON format as possible and ensure topic of interest is always mentioned in the question from this text:'
                prompt = f'{prompt_prefix} {text}'
                response = get_response(api_key, prompt, max_tokens=None, temperature=0.7, model="gpt-3.5-turbo")
                qa_pairs = json.loads(response.choices[0].message.content)  # a list of dictionaries; each dictionary contains a set of QA pair

                # export and save generated json question-answer pairs
                savefilename = file.replace('.txt','')
                
                os.makedirs(f'{savedata_dir}/{folder}', exist_ok=True)
                with open(f'{savedata_dir}/{folder}/{savefilename}_qa.json', 'w', encoding='utf-8') as f:
                    json.dump(qa_pairs, f, ensure_ascii=False, indent=4)

                f.close()
                
                # remove file after generating QA pair JSON successfully
                os.remove(f'{rawdata_dir}/{folder}/{file}')

                counter += 1
            
        except Exception as ex:
            # if error, move text file to error reprocess folder for manual fixing
            os.makedirs(f'{error_dir}/{folder}', exist_ok=True)
            os.rename(f'{rawdata_dir}/{folder}/{file}', f'{error_dir}/{folder}/{file}')
            
            if 'rate limit reached' in str(ex).lower():
                print(f'Error (at {folder}/{file}): {ex}')
                break_ind = True
                break
            else:
                pass
            
    if break_ind == True:
        break
```

## Consolidate and split all generated QA pairs into datasets for fine-tuning

```python
curr_dir = os.getcwd()
orig_dir = f'/Users/dahliama/Desktop/data298/github/projectmarley/data/data-preprocess/02_data_prompts/all_data'
print(f'Current file path: {curr_dir}')
```

```python
# get a list of text files from directory generated with original code
orig_files_list = []
for fold in [folder for folder in os.listdir(orig_dir) if '.' not in folder]:
    files = [f for f in os.listdir(f'{orig_dir}/{fold}') if '.json' in f]
    orig_files_list += files

print(f'There is a total of {len(orig_files_list)} question-answer json files curated.')
```

**Combine and split all data into train, validate and test sets**

```python
# combine all data into one list -- a list of dictionaries where each dictionary contains a question-answer pair
all_data = []
for fold in [folder for folder in os.listdir(orig_dir) if '.' not in folder]:
    files = [f for f in os.listdir(f'{orig_dir}/{fold}') if '.json' in f]
    
    print(f'Consolidating files from {fold} folder')
    
    for i in tqdm(range(len(files))):
        file = files[i]
        file_obj = open(f'{orig_dir}/{fold}/{file}', 'r')
        json_data = json.load(file_obj)
        all_data += json_data

print(f'There is a total of {len(all_data)} question-answer pairs consolidated.')
```

```python
# save the consolidated data
savefilename = 'all_data'
with open(f'{orig_dir}/{savefilename}.json', 'w', encoding='utf-8') as f:
    json.dump(all_data, f, ensure_ascii=False, indent=4)
```

**split data to train-validate-test sets**
* train = 70%
* validate = 15%
* test = 15%

```python
# set split sizes
total_size = len(all_data)
train_size = int(np.round(total_size*0.7))
val_size = int(np.round(total_size*0.15))
test_size = int(np.round(total_size*0.15))

if (train_size+val_size+test_size) == total_size:
    print(f'Splitting is good: train = {train_size}, val = {val_size}, test = {test_size}')
else:
    print(f'Splitting is not good: train = {train_size}, val = {val_size}, test = {test_size}')
```

```python
# set seed to always generate the same random numbers
random.seed(5)

# split data
ixs = list(range(total_size)) # get indices

train_ixs = random.sample(ixs, train_size)
train_data = [all_data[i] for i in train_ixs]

ixs_remain = [ix for ix in ixs if (ix not in train_ixs)]
val_ixs = random.sample(ixs_remain, val_size)
val_data = [all_data[i] for i in val_ixs]

ixs_remain = [ix for ix in ixs if (ix not in train_ixs) and (ix not in val_ixs)]
test_ixs = random.sample(ixs_remain, test_size)
test_data = [all_data[i] for i in test_ixs]

print(f'Train data size = {len(train_data)}')
print(f'Val data size = {len(val_data)}')
print(f'Test data size = {len(test_data)}')
print(f'Total data = {len(train_data)+len(val_data)+len(test_data)}')
```

```python
# save data splits
savefilenames = ['train_data','val_data','test_data']
saveobjs = [train_data, val_data, test_data]

for i in range(len(savefilenames)):
    savefilename = savefilenames[i]
    saveobj = saveobjs[i]
    
    with open(f'{orig_dir}/{savefilename}.json', 'w', encoding='utf-8') as f:
        json.dump(saveobj, f, ensure_ascii=False, indent=4)
```
